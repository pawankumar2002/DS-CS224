```python
import pandas as pd
import numpy as np
from scipy.stats import norm,pearsonr
import io
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('Education and gender equality.csv')
df=df.loc[0:35]
df
```

```python
mean = 50.11
```

```python
df = df.sort_values('Education',ascending = False)
new_df =df.head(5)
new_df
```

```python
GndEq=df['Gender Equality']
from scipy.stats.stats import ttest_1samp
t_value,p_value = ttest_1samp(GndEq,mean)
print('Test statistic is %f'%float("{:.6f}".format(t_value)))

print('p-value for one tailed test is %f'%p_value)
```
